```python
# charge integration
# sometimes np.trapz integration takes long...
import matplotlib.pyplot as plt

for i in selxn:
    fig, fig2, fig3 = i.charge_int(
        lower_lim=-1.09, upper_lim=0.03, target_RE='SHE', atomic=True, base_cap=0.1, c_corr={})
fig.legend()
fig2.legend()
fig3.legend()
```
